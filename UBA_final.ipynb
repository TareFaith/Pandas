```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\USER\OneDrive\Documents\uba new\UBA_GENDER.xlsx")
df.head()
```

```python
df=df.rename(columns={"gender_id": "Gender", "district": "District"})
df.head()
```

```python
df2=df.groupby(["District", "Gender"]).size().reset_index(name="Count")
print(df2)
```

```python
df['District'].fillna('Not Specified', inplace=True)

df2=df.groupby(["District", "Gender"]).size().reset_index(name="count")

pivot_result= df2.pivot(index="District", columns="Gender", values="count").reset_index()

pivot_result.fillna(0, inplace=True)

pd.options.display.float_format = '{:.0f}'.format

pivot_result.columns.name = None

print(pivot_result)
```

```python
file_path = r"C:\Users\USER\OneDrive\Documents\uba new\Final_uba.csv"

# Export the DataFrame to a CSV file
pivot_result.to_csv(file_path, index=False)  # Set index=False to exclude the DataFrame index in the output

print(f'Data has been exported to {file_path}')
```
